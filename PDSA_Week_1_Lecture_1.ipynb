```python
def factors(n):
  factors = []
  for i in range(1, n+1):
    if n % i == 0:
      factors.append(i)
  return factors
```

```python
def prime(n):
  return (factors(n) != [1, n])
```

```python
def prime(n):
  return (len(factors(n)) == 2)
```

```python
primelist = [ ]
for i in range(1, 100):
  if prime(i):
    primelist.append(i)
print(primelist)
```
